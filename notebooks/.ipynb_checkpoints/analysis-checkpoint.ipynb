# TCR Repertoire Sequencing — Literature Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the abstracts
df = pd.read_csv('../data/abstracts.csv')

print(f"Total abstracts: {len(df)}")
print(f"Year range: {df['year'].min()} – {df['year'].max()}")
print(f"\nAbstracts per year:")
print(df['year'].value_counts().sort_index())
```

```python
# Plot abstracts per year
fig, ax = plt.subplots(figsize=(8, 5))
year_counts = df['year'].value_counts().sort_index()
ax.bar(year_counts.index.astype(str), year_counts.values, color='steelblue', edgecolor='white')
ax.set_title('PubMed Abstracts per Year\nQuery: TCR Repertoire Sequencing', fontsize=14)
ax.set_xlabel('Year')
ax.set_ylabel('Number of Abstracts')
for i, v in enumerate(year_counts.values):
    ax.text(i, v + 1, str(v), ha='center', fontsize=11)
plt.tight_layout()
plt.savefig('../data/abstracts_per_year.png', dpi=150)
plt.show()
print("Plot saved.")
```

```python
from collections import Counter
import re

# Combine all abstracts into one big string
all_text = ' '.join(df['abstract'].dropna().tolist()).lower()

# Remove common English stopwords and short words
stopwords = {
    'the', 'a', 'an', 'and', 'or', 'of', 'in', 'to', 'is', 'are', 'was',
    'were', 'for', 'with', 'this', 'that', 'these', 'those', 'we', 'our',
    'their', 'from', 'by', 'as', 'be', 'have', 'has', 'had', 'not', 'at',
    'on', 'also', 'which', 'but', 'it', 'its', 'than', 'may', 'can', 'cells',
    'cell', 'using', 'used', 'between', 'among', 'both', 'into', 'during',
    'however', 'been', 'high', 'low', 'significantly', 'including', 'showed'
}

# Tokenize and filter
words = re.findall(r'\b[a-z]{4,}\b', all_text)
filtered = [w for w in words if w not in stopwords]

# Count and display top 30
counter = Counter(filtered)
top_words = counter.most_common(30)

print("Top 30 most frequent words across all abstracts:\n")
for word, count in top_words:
    print(f"  {word:<25} {count}")
```

```python
# Plot top 20 words
words_list = [w for w, c in top_words[:20]]
counts_list = [c for w, c in top_words[:20]]

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(words_list[::-1], counts_list[::-1], color='steelblue', edgecolor='white')
ax.set_title('Top 20 Most Frequent Words in TCR Repertoire Abstracts', fontsize=14)
ax.set_xlabel('Frequency')
plt.tight_layout()
plt.savefig('../data/top_words.png', dpi=150)
plt.show()
print("Plot saved.")
```
